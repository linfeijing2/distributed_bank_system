```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# filenames = ['node1_Bandwidth.txt', 'node2_Bandwidth.txt','node3_Bandwidth.txt']

# with open('BW.txt', 'w') as outfile:
#     for fname in filenames:
#         with open(fname) as infile:
#             outfile.write(infile.read())
            
```

```python
X = np.array([])
n=0
for file in ['node1_small_Bandwidth.txt', 'node2_small_Bandwidth.txt','node3_small_Bandwidth.txt']:
    n = n+1
    table = (pd.read_table(file, sep=' ', header=None, names=['bandWidth', 'Time'])).astype(float)
    table['Time']  = (table['Time']).apply(np.floor)
    table['Time']  = table['Time'] - table['Time'][0]

#     print(table)

    tableSum = table.groupby('Time').sum().reset_index()

    xAxis = tableSum['Time'].to_numpy()

    yAxis = tableSum['bandWidth'].to_numpy()
    plt.figure(figsize=(8, 6),dpi=110)
    plt.plot(xAxis, yAxis, '-b.')
    plt.title("node"+str(n)+"Bandwidth")
    plt.xlabel('Time (s)')
    plt.ylabel('Bytes')
    plt.legend()
    plt.savefig("graph/smallScenario/node"+str(n)+"Bandwidth"+".png")
    plt.show()
    
```

```python
filenames = ['node1_small_Processing.txt', 'node2_small_Processing.txt','node3_small_Processing.txt']

with open('3nodes_PT.txt', 'w') as outfile:
    for fname in filenames:
        with open(fname) as infile:
            outfile.write(infile.read())
X = np.array([])

table = (pd.read_table("3nodes_PT.txt", sep=' ', header=None, names=['msgID', 'Time']))
# table['Time']  = (table['Time']).apply(np.floor)
# table['Time']  = table['Time'] - table['Time'][0]


# print(table)

tableMin = table.groupby('msgID').min()
tableMax = table.groupby('msgID').max()
# print((tableMin))
# print((tableMax))

tableMax['msgIndex'] = range(1, len(tableMax) + 1)
xAxis = tableMax['msgIndex'].to_numpy()
print(len(xAxis))
yAxis = (tableMax['Time'] - tableMin['Time']).to_numpy()

# print(yAxis)

plt.figure(figsize=(8, 6),dpi=110)
plt.plot(xAxis, yAxis, '.b-')
plt.title("Message Process Time at All Nodes")
plt.xlabel('Message')
plt.ylabel('Delay Time (s)')
plt.savefig("graph/smallScenario/3 Nodes Processing Time.png")
# plt.show()
```
